```python
################################################### IMPORTS ##########################################################

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

#original database
concdata = pd.read_csv("Concrete_Data_Yeh_final.csv",)
print(f"Original concrete database:\n\n{concdata}\n")

################################################### PREPROCESSING #####################################################
################################################## 1. REMOVING NaN ####################################################

#replacing NaN with mean 
mean_values = {
    "cement": concdata["cement"].mean(),
    "slag": concdata["slag"].mean(),
    "flyash": concdata["flyash"].mean(),
    "water": concdata["water"].mean(),
    "superplasticizer": concdata["superplasticizer"].mean(),
    "coarseaggregate": concdata["coarseaggregate"].mean(),
    "fineaggregate": concdata["fineaggregate"].mean(),
    "age": concdata["age"].mean(),
    "csMPa": concdata["csMPa"].mean()
}

mean_values_df = pd.DataFrame(list(mean_values.items()), columns=['Attribute', 'Mean'])
print(f"Mean values: \n{mean_values_df}\n")
concdata_replaced_mean = concdata.copy().fillna(value=mean_values)
print(f"Mean values substituted: \n\n{concdata_replaced_mean}\n")

#replacing NaN with median
median_values = {
    "cement": concdata["cement"].median(),
    "slag": concdata["slag"].median(),
    "flyash": concdata["flyash"].median(),
    "water": concdata["water"].median(),
    "superplasticizer": concdata["superplasticizer"].median(),
    "coarseaggregate": concdata["coarseaggregate"].median(),
    "fineaggregate": concdata["fineaggregate"].median(),
    "age": concdata["age"].median(),
    "csMPa": concdata["csMPa"].median()
}

median_values_df = pd.DataFrame(list(median_values.items()), columns=['Attribute', 'Median'])
print(f"Median values: \n{median_values_df}\n")
concdata_replaced_median = concdata.copy().fillna(value=median_values)
print(f"Median values substituted: \n\n{concdata_replaced_median}\n")

################################################## 2. USING SCALERS ####################################################

# min-max scaler
min_max_scaler = MinMaxScaler()
concdata_min_max_scaled = min_max_scaler.fit_transform(concdata_replaced_mean)
concdata_min_max_scaled_df = pd.DataFrame(concdata_min_max_scaled, columns=concdata.columns)

print(f"MinMax Scaled data: \n{concdata_min_max_scaled_df}\n")

# standard scaler

standard_scaler = StandardScaler()
concdata_standard_scaled = standard_scaler.fit_transform(concdata_replaced_mean)
concdata_standard_scaled_df = pd.DataFrame(concdata_standard_scaled, columns=concdata.columns)

print(f"Standard Scaled data: \n{concdata_standard_scaled_df}\n")
```
